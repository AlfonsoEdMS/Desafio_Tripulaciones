```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Base de datos colegios.xlsx - Colegios.csv')
# df
```

```python
work = df.copy()
work.info()
```

Borrar solo las que no tengan ni email ni telefono!!

```python
# Nos deshacemos de aquellos que no tienen Email
work = work.dropna(subset=['Email'])
work = work.dropna(subset=['Telefono'])

# Nos deshacemos de la columna FAX
work = work.drop(['Fax', 'Pais'], axis=1)

# Separmos la última columna en Categoría y Tipo
new_columns = work["Categoria"].str.split('|', expand=True)
new_columns.columns = ['Categoria', 'Tipo']

# Eliminamos la columna 'Categoría' y Concatenamos df's
work = work.drop(['Categoria'], axis=1)
work = pd.concat([work, new_columns], axis=1)

# Reset
work.reset_index(drop = True, inplace=True)
```

### Columna Tipo

```python
work['Tipo'].nunique(), len(work)
```

```python
# 1/
list_of_types = ['Danza', 'Múisca', 'Adultas', 'Universidades', 'Conservatorio', 'Profesional', 'Superior', 
'Festivales', 'Hostelería', 'Enseñanza', 'Academias']
# Tenemos que deshacernos de todas aquellas Academías, etc (Centros educaivos no relativos a nuestros intereses)
list_of_erase_values = []
for palabra in list_of_types:
    for index, registro in enumerate(work['Categoria']):
        if palabra in registro:
            list_of_erase_values.append(index)
        
print(len(list_of_erase_values))
# 2/
work.drop(work.index[list_of_erase_values], inplace = True )
# 3/
work.reset_index(drop = True, inplace=True)
```

```python
work['Tipo'].nunique(), len(work)
```

```python
# Segunda Manera de hacerlo
DF1 = work[work.Tipo.str.contains(
    'Conservatorio|Múisca|Adultas|Universidades|Profesional|Superior|Festivales|Hostelería|Enseñanza|Academias'
, na=False)]
len(DF1)
```

```python
# carrying out anti join using merge method
df3 = work.merge(DF1, on='Nombre', how='left', indicator=True)
  
df = df3.loc[df3['_merge'] == 'left_only', 'Nombre']
  
work_1 = work[work['Nombre'].isin(df)]
```

### Siguiente Columna (Categoría)

```python
# 1/
list_of_categories = ['Academias', 'Universidades', 'Adultos', 'Profesores', 'Psicología ', 'Profesional', 'terapeutas', 
'Adultos', 'Organizaciones', 'Autoescuelas','Artes', 'ACADEMIAS', 'Terapias', 'Obras', 'Superiores', 'Organismos', 
'Deportes', 'profesional', 'Acupunt', 'Aeródromos', 'UNED''Universitario', 'Traducciones', 'Agencias', 'musical', 
'Formación', 'orientación', 'Fiestas', 'Música', 'Heraldica', 'Fotografia', 'Pilotos', 'Informática', 'Deportiva', 'Idiomas', 'Literatura', 'rehabilitación', 'ESTETICA', 
'Higiene', 'Canina', 'Fundaciones']
# Tenemos que deshacernos de todas aquellas Academías, etc (Centros educaivos no relativos a nuestros intereses)
list_of_erase_values = []
for palabras in list_of_categories:
    for index, registro in enumerate(work_1['Categoria']):
        if palabras in registro:
            list_of_erase_values.append(index)
print(len(list_of_erase_values))
# 2/
work_1.drop(work_1.index[list_of_erase_values], inplace = True )
# 3/
work_1.reset_index(drop = True, inplace=True)
```

```python
work['Categoria'].nunique(), len(work)
```

```python
# Segunda Manera de hacerlo
DF2 = work_1[work_1.Tipo.str.contains(
    'Academias|Universidades|Adultos|Profesores|Psicología|Profesional|terapeutas|Adultos|Organizaciones|Autoescuelas|Artes|ACADEMIAS|Terapias|Obras|Superiores|Organismos|Deportes|profesional|Acupunt|Aeródromos|UNED|Universitario|Traducciones|Agencias|musical|Formación'
, na=False)]

# carrying out anti join using merge method
df2 = work_1.merge(DF2, on='Nombre', how='left', indicator=True)
  
df = df2.loc[df2['_merge'] == 'left_only', 'Nombre']
  
work_2 = work_1[work_1['Nombre'].isin(df)]
```
